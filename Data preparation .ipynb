```python
def get_info_fleurs_parc(df, nomparc):
    All_species = []
    All_genders = []
    All_families = []
    subdf = df[df['Nom'] == nomparc]
    subdf = subdf.reset_index(drop=True)
    for line in range(len(subdf)):
        All_species.append(subdf.loc[line, 'Espèce'])
        All_genders.append(subdf.loc[line, 'Genre'])
        All_families.append(subdf.loc[line, 'Famille'])
    
    unique_Species = list(set(All_species))
    unique_Genders = list(set(All_genders))
    unique_Families = list(set(All_families))
    
    print("{} dispose de {} plantes fleuries, disposées en {} familles de {} genres pour un toltal de {} espèces".format(nomparc,len(df),len(unique_Families),len(unique_Genders),len(unique_Species)))

get_info_fleurs_parc(CVN, 'PLANTES')
get_info_fleurs_parc(CVN, 'PROCES')
get_info_fleurs_parc(CVN, 'BEAUJOIRE')
get_info_fleurs_parc(CVN, 'BLOTTEREAU')
get_info_fleurs_parc(CVN, 'GAUDINIERE')
```
